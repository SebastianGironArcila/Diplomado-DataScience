### Subgráficas

```python
fig = plt.figure()


ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)

ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

```python
fig4 = plt.figure()

ax1_4 = fig4.add_subplot(2, 2, 1) # first position
ax2_4 = fig4.add_subplot(2, 2, 2) # second position
ax3_4 = fig4.add_subplot(2, 2, 3) # third position

plt.plot(np.random.randn(50).cumsum(), 'k--')

_ = ax1_4.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)

ax2_4.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

#### Ajustando espacio entre gráficas

```python
# we will make two subplots with the same x- and y-axis 
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()

# we will make two subplots with the same x- and y-axis 
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=1, hspace=1)
```

### Colores, marcadores y estilos de linea
Cada subgráfica de Matplotlib acepta arreglos de coordenadas Y y X y opcionalmente un string de abreviacion para el color y estilo de la linea.
Ej:

`ax.plot(x, y, 'g--')`

Así construimos una gráfica de X vs Y con linea punteada verde. Esta es una sintaxis alternativa:

`ax.plot(x, y, linestyle='--', color='g')`

```python
import numpy as np

plt.plot(np.random.randn(30).cumsum(), color='k', linestyle='--', marker='o')
```

### Ubicación de linea
image.png

```python
import numpy as np

figure = plt.figure()

data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')

figure.legend(loc='best')
figure.suptitle("Post")
```

```python
import numpy as np

figure = plt.figure()

data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-pre', label='steps-pre')

figure.legend(loc='best')
figure.suptitle("Pre")
```

### Personalizar ejes

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)

# Configuración por defecto
ax.plot(np.random.randn(1000).cumsum())
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
# The rotation option sets the x tick labels at a 30-degree rotation.
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')

ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```
